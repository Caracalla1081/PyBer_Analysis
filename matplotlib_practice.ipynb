```python
%matplotlib inline
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "June", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.01, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plot()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with an ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label="Boston")
# Create labels for the x and y axis
plt.xlabel("date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0,45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a gird.
plt.grid()
# Add the legend.
plt.legend()
```

```python
fig, ax = plt.subplots()
# Create the plot and add a label for the legend.
ax.plot(x_axis, y_axis, color="green", marker="^", linewidth=2, label="line")
# Create labels for the x and y axis
ax.set_xlabel("date")
ax.set_ylabel("Fare($)")
# Set the y limit between 0 and 45.
ax.set_ylim(0,45)
# Create a title.
ax.set_title("PyBer Fare by Month")
# Add a gird.
ax.grid()
# Add the legend.
ax.legend("Boston")
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis, y_axis)
```

```python
# Create the opposite axes plot from cell above.
plt.barh(y_axis, x_axis)
```

```python
# Create the plot, inverting the month's axis.
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
# Create the plot, inverting the month's axis and formatting changes.
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
plt.gca().invert_yaxis()
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create the plot with ax.plt as horizontal bar chart.
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
#Create barh with reverse axes.
fig, ax =plt.subplots()
ax.barh(x_axis, y_axis, color = "cyan")
ax.legend = "Chicago"
ax.set_title("PyBer Fare by Month")
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")
ax.invert_yaxis()
```

```python
# Create scatter plot
plt.plot(x_axis, y_axis, 'o')
```

```python
# Create scatter plot alt method
plt.scatter(y_axis, x_axis, color="red", label="Chicago")
plt.xlim(0,50)
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.legend()
plt.gca().invert_yaxis()
```

```python
# Create Bubble Chart
plt.scatter(x_axis, y_axis, s= [i*3 for i in y_axis])
```

```python
# Create Scatter Plot w/Object oriented method
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create Bubble Chart w/Obj oriented method
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s= [i*5 for i in y_axis], color="skyblue", linewidth=2, edgecolors="black", alpha=0.2)
ax.legend("Boston")
ax.set_xlim(0,50)
ax.invert_yaxis()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
```

```python
# Create pie chart with matlab
plt.pie(y_axis, labels=x_axis)
plt.show
```

```python
# Create pie chart with further detail
plt.subplots(figsize=(9,9))
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.subplots(figsize=(9,9))
plt.pie(y_axis, explode=explode_values, colors=colors, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)
fig, ax = plt.subplots(figsize=(9,9))
ax.pie(y_axis, labels=x_axis, autopct='%.1f%%', explode=explode_values, colors=colors, shadow=1, startangle=95, counterclock=False)
plt.show
```
